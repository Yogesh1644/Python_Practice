```python
# creating book dataframe
```

```python
book_name=[]
auther_name=[]
price=[]
sale_price=[]
```

```python
book_dict={}
```

```python
books=soup.find_all('product-grid-item',class_="product-grid-item group/product-grid-item")
for b in books:
    book_name.append(b["aria-label"])
    #print(b.find('div',class_="product__grid__info text-left").find('p',class_="author-name"))
    if str(b.find('div',class_="product__grid__info text-left").find('p',class_="author-name")) =='None':
        auther_name.append(b["aria-label"])
    else:
        auther_name.append(b.find('div',class_="product__grid__info text-left").find('p',class_="author-name").string.strip().split(' ',1)[1]) 

    price.append(b.find('span',class_="compare-at").string.strip().split(' ')[1])
    sale_price.append(b.find('span',class_="price on-sale").string.strip().split(' ')[1])
book_dict['Book']=book_name
book_dict['Auther']=auther_name
book_dict['Price']=price
book_dict['Sale_Price']=sale_price

 
    
   # print('y')
    
```

```python
book_dict
```

```python
bdf=pd.DataFrame(book_dict)
```

```python
bdf
```

```python
bdf['Discount %']=((1-bdf['Sale_Price'].astype(float)/bdf['Price'].astype(float))*100).round(2)
```
